# IdiomGen: Phrase Generator

This notebook focuses on the collection and generation of American idioms. This is accomplished through 
* the use of a custom webscraper that pulls all idioms and their first definition from http://theidioms.com. 
* building of a Markov chain through the Markovify package based upon said idioms.
* and the writing of new idioms generated by this model to a text file.


Import necessary packages

```python
import pandas as pd
import markovify
from Webscraper import idiom_scraper
```

Deploy Webscraper

```python
idiom_scraper('idiom_list.csv', page_count=152, limit=145, timer=True)
```

Convert csv to txt for ease of use with Markovify

```python
df_idioms =  pd.read_csv('idiom_list.csv', encoding='latin1')
df_idioms['PHRASE'].to_csv('idiom_phrase.txt', sep="\n", index=False, header=False)

with open('idiom_phrase.txt') as f:
    text = f.read()
f.close()
```

Build our model and generate new idioms based upon the data scraped from http://theidioms.com

```python
idiom_model = markovify.NewlineText(text, state_size=2, well_formed=False)

def generate_idioms(n):
    with open('new_idioms.txt', 'w') as f:
        for i in range(n):
            f.write(idiom_model.make_short_sentence(140) + "\n")
    f.close()
```

```python
generate_idioms(10)
```
